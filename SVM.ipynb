```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

data=pd.read_csv("imdb.csv")
print("Dataset : \n",data)

print("Null values : \n" ,data.isnull().sum())

#imputation
rev_mean = data['Revenue (Millions)'].mean()
print("\nMean of 'Revenue(Millions)' : ",rev_mean)
data['Revenue (Millions)'].fillna(rev_mean,inplace=True)
met_mean = data['Metascore'].mean()
print("Mean of 'Metascore' : ",met_mean)
data['Metascore'].fillna(met_mean,inplace=True)
print("\nNull values after imputation : \n" ,data.isnull().sum())

bins=[0,50,100]
data['Metascore'] = pd.cut(data['Metascore'],bins=bins)
label_encoder = LabelEncoder()
data['Metascore'] = label_encoder.fit_transform(data['Metascore'])

x=data.iloc[:,[1,2,3,5]]
y=data.iloc[:,-1]

print("\nIndependent variables are : \n",x)
print("\nDependent variables are : \n",y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)

mod=SVC(kernel='linear', random_state=0) 
model=mod.fit(x_train,y_train)
print("\nModel : ",model)
y_pred=model.predict(x_test)
print(f'\ny_test : \n{y_test} \n\ny_pred : \n{y_pred}')
print("\nConfusion Matrix : \n",confusion_matrix(y_pred,y_test))
print("\nAccuracy Score : ",accuracy_score(y_pred,y_test))
```
