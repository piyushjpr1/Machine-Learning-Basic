```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,confusion_matrix,accuracy_score,classification_report

Dataset=pd.read_csv("companies.csv")
print("Dataset : \n",Dataset)
x=Dataset.iloc[:,:2]
y=Dataset.iloc[:,-1]
print("Independent variables are : \n",x)
print("Dependent variables are : \n",y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=4)
lm = LinearRegression()
model=lm.fit(x_train,y_train)
print("Model : \n",model)
y_pred=model.predict(x_test)
print(f'\ny_test = {y_test} \ny_pred : {y_pred}')
print("Coefficients : \n",model.coef_)
print("\nMean squared error : %.2f " % mean_squared_error(y_test,y_pred))
print("\nCoefficient of determination : %.2f " % r2_score(y_test,y_pred))
plt.scatter(Dataset['R&D Spend'],Dataset['Profit'],color='black')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.show()
```
