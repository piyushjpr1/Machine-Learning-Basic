```python
import numpy as np
import pandas as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import matplotlib.pyplot as plt

dataset = datasets.load_iris()
print("Dataset : \n",dataset)

x=dataset.data
y=dataset.target
print("\nIndependent variables are : \n",x)
print("\nDependent variables are : \n",y)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)

mod=KNeighborsClassifier(3)
model=mod.fit(x_train,y_train)
print("\nModel : ",model)

y_pred=model.predict(x_test)
print(f'\ny_test : \n{y_test} \n\ny_pred : \n{y_pred}')
print("\nConfusion Matrix : \n",confusion_matrix(y_test,y_pred))
print("\nAccuracy Score : ",accuracy_score(y_pred,y_test))
print("\nclassification_report : \n",classification_report(y_test,y_pred))

plt.scatter(x[51:100,2],x[51:100,3],color="black")
plt.scatter(x[1:50,2],x[1:50,3],color="green")
plt.scatter(x[101:150,2],x[101:150,3],color="pink")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
