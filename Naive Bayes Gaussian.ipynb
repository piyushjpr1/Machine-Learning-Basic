```python
# Import models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score

# read dataset
Dataset=pd.read_csv("heart.csv")
print("Dataset : \n",Dataset)

# Distinguish dependent and independent variables
x=Dataset.iloc[:,:12]
y=Dataset.iloc[:,-1]
print("Independent variables are : \n",x)
print("Dependent variables are : \n",y)

# split test and training data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=4)

# fit the model
mod=GaussianNB()
model=mod.fit(x_train,y_train)
print("\nModel : ",model)

# Prediction
y_pred=model.predict(x_test)
print(f'\ny_test : \n{y_test} \n\ny_pred : \n{y_pred}')
print("\nConfusion Matrix : \n",confusion_matrix(y_pred,y_test))
print("\nAccuracy Score : ",accuracy_score(y_pred,y_test))
```
